# 2. Simulacion parametros Z


Para modelar un transformador en LTSpice se puede utilizar inductores acoplados. 

<div>
    <img src="inductores.png" width="200"/>
</div>

Un par de inductores acoplados con bornes no homologos como el que se muestra en la imagen tiene la siguiente relacion de tensiones con corrientes:

- $V_1 = L_1s I_1 - M I_2$
- $V_2 = L_2s I_2 - M I_1$

(Corrientes entrantes en los bornes superiores)

Donde $M = k \sqrt{L_1 L_2}$ y $k$ es la constante de acoplamiento.

Eligiendo $L = L_1 = L_2$ y $k = 1$ las ecuaciones se reducen a:


- $V_1 = Ls (I_1 - I_2)$
- $V_2 = Ls (I_2 - I_1)$

Calculando la relación de tensiones obtenemos:

$\dfrac{V_1}{V_2} = \dfrac{I_1 - I_2}{I_2 - I_1} = -1$

$V_1 = - V_2$

Calculando la relación de corrientes obtenemos:

$\dfrac{I_1}{I_2} = \dfrac{\dfrac{V_1}{Ls} + I_2}{\dfrac{V_2}{Ls} + I_1}$

En el caso donde $Ls$ tiende a infinito: ($Ls \to +\infty$)

$\dfrac{I_1}{I_2} = \dfrac{I_2}{I_1}$

$I_1 = I_2$

Estas relaciones coinciden con las de un transformador ideal.

Concluimos que se puede aproximar un transformador ideal con relación de transformacion unitaria trabajando en frecuencias altas, utilizando un transformador con inductancias altas, eligiendo $k=1$ y utilizando inductores de igual valor para el primario y el secundario.

$$T(s) = \dfrac{V_o}{V_i} = \dfrac{1}{\dfrac{L_1 L_3 C_2}{R}s^3 + L_1 C_2 s^2 + \left(\dfrac{L_1 + L_3}{R} \right) s + 1}$$
